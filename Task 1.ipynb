#Loading the dataset and checking for inconsistencies.

```python
import pandas as pd
df = pd.read_csv("Cleaned_DS_Jobs.csv")
print(df.info()) # Check for missing values and data types
```

#Identify and remove redundant entries.

```python
df = df.drop_duplicates() # Removes exact duplicate rows [cite: 8, 28]
```

#Diagnose missing data and apply imputation or deletion.


```python
df["Rating"] = df["Rating"].fillna(df["Rating"].mean()) # Imputation
```

#Fix numerical columns trapped as strings and normalize text.

```python
df["Job Title"] = df["Job Title"].str.lower().str.strip() # Text normalization [cite: 9, 35]
df["Salary Estimate"] = pd.to_numeric(df["Salary Estimate"], errors='coerce') # Type correction
```

Using assertions to ensure no missing values or duplicates remain before saving.

```python
assert df.duplicated().sum() == 0 # Ensure no duplicates 
df.to_csv("cleaned_data.csv", index=False) # Save final output [cite: 37]
```
